# Ejercicio

Una empresa de logística cobra a sus clientes según la distancia que recorre el paquete. La tarifa se rige según la siguiente tabla:

| Distancia| Precio|
|----------|-----------------|
| 0 a 9.99 Km | 2 USD por Km|
| 10 a 49.99 Km | 1.8 USD por Km|
| 50 a 99.99 Km | 1.5 USD por Km|
| 100 a 249.99 Km | 1.25 USD por Km|
| 250 Km o más | 1 USD por Km|

Adicionalmente, existe un recargo sobre el total a pagar que depende del peso del paquete. Este recargo se rige según la siguiente tabla:

|Peso|Recargo|
|----------|-----------------|
|0 a 99.99 Kg|Sin recargo adicional|
|100 a 499.99 Kg| 2% de recargo|
|500 Kg o más| 5% de recargo|

También la empresa cuenta con un sistema de fidelidad para otorgar beneficios a sus clientes. Existen 3 niveles de fidelidad:
- Nivel 1: Es el nivel básico de todos los clientes. No tiene beneficios adicionales
- Nivel 2: Otorga un descuento del 3% sobre el total a pagar
- Nivel 3: Otorga un descuento del 5% sobre el total a pagar. Adicionalmente elimina los recargos por peso.

Un cliente puede pagar varios envíos en una misma factura de cobro. Elabore un programa que genere una factura de cobro: primero, lea el nivel de fidelidad del cliente. Luego, que ingrese la cantidad envíos que desea cotizar. Después, para cada envío, lea la distancia (en Km) y el peso (en Kg). Al final, muestre el total a pagar por todos los envíos cotizados.
